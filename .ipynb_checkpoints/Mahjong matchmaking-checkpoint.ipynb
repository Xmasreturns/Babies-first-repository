```python
""" This program takes input from the scores database table (currently csv, later sql).
 Given a list of player IDs, it generates tables that attempt to make 4 or 5
 player tables of people who have played each other the least."""
#Uploaded to Github 8/10/15

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from __future__ import division
import itertools
from itertools import combinations
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# todays_players can't have duplicates
todays_players = [40,6,23,26,15,
                  98,1,10,55,16,
                  56,97,96,12,13,
                  41,52,88,66,94,
                  89,86,87,77,79,
                  72,76]
```

```python
scores_table = pd.read_csv('MahjongScores.csv')
```

```python
scores_table.head()
```

```python
scores_table.info()
```

```python
#See what playerID looks like
print scores_table['PlayerId'].min()
print scores_table['PlayerId'].max()
```

```python
# 6th column is extra, drop it
scores_table.drop([scores_table.columns[6]], axis=1, inplace=True) # Note: zero indexed
scores_table.head()
```

```python
games_rows = pd.DataFrame()

# Pull all rows with games for each of today's players
for i in todays_players:
    games_rows = games_rows.append(scores_table.loc[(scores_table['PlayerId'] == i)])

# Make a list of just unique games
uniquegames = games_rows['GameId'].unique()
```

```python
# Create zeros dataframe of (#players,#players) size for matchups
num_players = len(todays_players)
matchups_df = DataFrame(np.zeros(num_players**2).reshape(num_players,num_players))
matchups_df.index = todays_players
matchups_df.columns = todays_players

# Populate matchups table one game at a time
for gameid in uniquegames:
    gameid_players = []
    relevant_rows = games_rows.loc[(games_rows['GameId'] == gameid)] # Pull the row for each player in that game
    gameid_players = relevant_rows['PlayerId'].tolist() #Give each player an index
    for i in range(0,len(gameid_players)): # For each player in that game
        for j in range(i,len(gameid_players)): # Look for opponent (may be self)
            matchups_df.loc[gameid_players[i]].loc[gameid_players[j]] += 1 # Add 1 to the column that player is in
                
# Diagonals should be total # of games played for that player
matchups_df
```

```python
# Determine amount of 5 person tables
if num_players < 8:
    total_tables = 1
    print "Everybody's at the same table"
else:
    tables_5p = num_players % 4
    total_tables = int(num_players / 4) #Assuming this rounds down
    tables_4p = total_tables - tables_5p
    print 'Total tables: %s' %total_tables
    print 'Five player tables: %s' %tables_5p
```

```python
def table_score(table):
    table_sum = 0
    for i in table:
        for j in table:
            if i>j: # This only serves to prevent repeats
                table_sum += max(matchups_df.loc[i].loc[j],matchups_df.loc[j].loc[i])
    return table_sum
```

```python
output_tables = []
tables_5p_made = 0
temp_playerlist = list(todays_players) # Make a copy of today's players to use

import datetime # Time it

# Make tables until all tables are made
while len(output_tables) < total_tables:
    
    # Make final table if only 4 players are in pool
    if (len(temp_playerlist) == 4):
        output_tables.append(tuple(temp_playerlist))
        break
    
    # Generate list of combinations for table size
    if tables_5p_made < tables_5p:
        table_combs = list(itertools.combinations(temp_playerlist,5))
        tables_5p_made += 1
    else:
        table_combs = list(itertools.combinations(temp_playerlist,4))

    startTime = datetime.datetime.now()
        
    # Calculate matchmaking rating for each table (lower means players have fewer games together)
    table_df = DataFrame(table_combs)
    table_sums = table_df.apply(table_score,axis=1)    
        
    endTime = datetime.datetime.now()
    print endTime-startTime
    
    # Take first minimum value in table_sums - gives priority to lower PlayerId
    best_table_ix = np.argmin(table_sums)
    print table_sums[best_table_ix]
    output_tables.append(table_combs[best_table_ix])
    
    # Remove players that have already been seated from the pool
    for player in table_combs[best_table_ix]:
        temp_playerlist.remove(player)
```

#Data Analysis
Now, let's look at aggregate scores and make some visualizations of skill.

First, assume we're only interested in the people playing today.

Some ways we could visualize this are:

    mean score vs opponent (matchup table)
    mean score over time (games)
    average rank and deviation (1st-5th)

###Mean score vs opponent

```python
# The data we'll need is all in games_rows for players in todays_players
games_rows.head()
```

```python
# The goal is to make a new matchup table for visualization, populating with mean scores
# Since we already have matchups_df for game count, we can sum the mean scores and divide
num_players = len(todays_players)
MUscores_df = DataFrame(np.zeros(num_players**2).reshape(num_players,num_players))
MUscores_df.index = todays_players
MUscores_df.columns = todays_players

for gameid in games_rows['GameId'].unique():
    gameid_players = []
    relevant_rows = games_rows.loc[(games_rows['GameId'] == gameid)] # Pull the row for each player in that game
    gameid_players = relevant_rows['PlayerId'].tolist() # Give each player in the game an index
    gameid_scores = relevant_rows['Score'].tolist() # Give each score in the game the same index
    for i in range(0,len(gameid_players)): # For each player in that game
        flag = False
        for j in range(0,len(gameid_players)): # Look for opponent (may be self)
            if (i == j):
                if flag == False: # Don't add along diagonal twice
                    MUscores_df.at[gameid_players[i],gameid_players[j]] += gameid_scores[i]
                flag = True
            else:
                MUscores_df.at[gameid_players[i],gameid_players[j]] += gameid_scores[i] # Add score of i to matchup sum
MUscores_df
```

```python
# Convert all sums into means (Diagonal is now player's average score, will confirm below)
for i in range(0,num_players):
    for j in range(0,num_players):
        MUscores_df.iloc[i,j] = MUscores_df.iloc[i,j] / np.diagonal(matchups_df)[j]
```

```python
MUscores_df.sort_index(axis=1, inplace=True)
MUscores_df.sort_index(axis=0, inplace=True)
MUscores_df
```

```python
# Make a red-green colormap
x = np.arange(25).reshape(5, 5)
cmap = sns.diverging_palette(10, 150, s=99, l=55, n=80, center='light',as_cmap=True)
```

```python
# Plot a heatmap of matchups, where diagonal is player personal average
plt.figure(figsize=(15,10))
ax = sns.heatmap(MUscores_df, cmap=cmap, square=True, vmin=-45, vmax=45)
ax.xaxis.tick_top()
```

###Scores over time

```python
# Making a df of scores over games for all relevant players
relevant_scores = pd.DataFrame()
relevant_scores['GameId'] = np.sort(games_rows['GameId'].unique())
for playerid in todays_players:
    relevant_scores[playerid] = np.nan

# Group games_rows by GameId and PlayerId to access multiple row values at once
grouped_scores = games_rows['Score'].groupby([games_rows['GameId'],games_rows['PlayerId']])

# Populate columns with each player's scores
for (game, player), score in grouped_scores:
    relevant_scores[player].loc[relevant_scores['GameId'] == game] = score.iloc[0]

relevant_scores.head()
```

```python
# Check table means to confirm earlier statement that the diagonal of MUscores is the mean:
relevant_scores.describe()
```

```python
relevant_scores.plot(x='GameId',figsize=(15,6),cmap='rainbow').legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

We see a lot of gaps due to players not being in every game

So, let's plot the moving average instead to better smooth things out

```python
moving_mean = pd.DataFrame()
moving_mean['GameId'] = relevant_scores['GameId']

for player in todays_players:
    moving_mean[player] = pd.rolling_mean(relevant_scores[player],window=100,min_periods=2)

moving_mean.loc[:,todays_players].plot(figsize=(15,6),cmap='rainbow').legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

###Average rank at table and deviation

```python
# Making a df of ranks over games for all relevant players
relevant_ranks = pd.DataFrame()
relevant_ranks['GameId'] = np.sort(games_rows['GameId'].unique())
for playerid in todays_players:
    relevant_ranks[playerid] = np.nan

# Group games_rows by GameId and PlayerId to access multiple row values at once
grouped_ranks = games_rows['Rank'].groupby([games_rows['GameId'],games_rows['PlayerId']])

# Populate columns with each player's scores
for (game, player), rank in grouped_ranks:
    relevant_ranks[player].loc[relevant_ranks['GameId'] == game] = rank.iloc[0]

relevant_ranks.head()
```

```python
rank_info = relevant_ranks.describe().fillna(0)
rank_info
```

```python
# Grab the means without GameId
rank_mean = rank_info.loc['mean'].iloc[1:]

# Sort the columns by mean to make plot look a bit nicer
rank_mean.sort(inplace=True)

# rank_info with columns sorted by mean values
rank_info = rank_info[rank_mean.index]
```

```python
rank_info.loc['mean'].plot(yerr=rank_info.loc['std'],kind='bar', figsize=(15,6))
```

Aside from the bars with no error windows (not enough games), we see that the leftmost players in the plot have the highest average ranking
